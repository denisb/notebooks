# Concat vs Join

Comparaison entre divers programmes, divers langages

```python
%%writefile concatVsJoin.pl
lireListe([E|L]):-
  readln([E|_]), !,
  lireListe(L).
lireListe([]).

ecrireListe([]).
ecrireListe([E|L]):-
  writeln(E),
  ecrireListe(L).

concat([],L,L).
concat([E|L],F,[E|R]):-
    concat(L,F,R).
    
main :- 
  writeln('Votre liste ? (un entier par ligne, un ligne blanche à la fin)'),
  lireListe(Entree1),
  writeln('Votre liste ? (un entier par ligne, un ligne blanche à la fin)'),
  lireListe(Entree2),
  concat(Entree1,Entree2,Sortie),
  writeln('Liste concaténée : '),
  ecrireListe(Sortie).

:-main.
```
