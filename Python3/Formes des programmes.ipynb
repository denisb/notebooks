# Formes des programmes

## La Fusion (ou interclassement) dans le tri par fusion

### algo classique

```python
def fusion(A,B):
    R=[]
    while A!=[] and B!=[]:
        if A[0]<B[0]:
            R=R+[A[0]]
            A=A[1:]
        else:
            R=R+[B[0]]
            B=B[1:]
    if A!=[]:
        R=R+A
    else:
        R=R+B
    return R

def triFusion(T): 
    if len(T)>1:
        T=fusion(triFusion(T[0:len(T)//2]),triFusion(T[len(T)//2:]))
    return T
            
```

```python
triFusion([3,1,4,1,5,9,2,6,5])
```

```python
import random
N=11111
triFusion([random.randint(0,N) for i in range(N)])
```

### Sans else ?

Est-ce que l'on peut éviter les "else" ? (quelle étrange question)

```python
def fusionSansElse(A,B):
    R=[]
    while A!=[] and B!=[]:
        if A[0]<B[0]:
            R=R+[A[0]]
            A=A[1:]
            continue
        R=R+[B[0]]
        B=B[1:]
    if A!=[]:
        return R+A
    return R+B

def triFusionSansElse(T): 
    if len(T)>1:
        T=fusionSansElse(triFusionSansElse(T[0:len(T)//2]),triFusionSansElse(T[len(T)//2:]))
    return T
```

```python
triFusionSansElse([3,1,4,1,5,9,2,6,5])
```

```python
import random
N=11111
triFusionSansElse([random.randint(0,N) for i in range(N)])
```

### Sans if ?

Est-ce que l'on peut enlever les "if" (dans la fusion) (quelle étrange question)

```python
def minMax(A,B):
    if A[0]<B[0]:
        return (A,B)
    return (B,A)
    
def fusionSansIfElse(A,B):
    R=[]
    while A!=[] and B!=[]:
        (A,B)=minMax(A,B)
        R=R+[A[0]]
        A=A[1:]
    return R+A+B

def triFusionSansIfElse(T): 
    if len(T)>1:
        T=fusionSansIfElse(triFusionSansIfElse(T[0:len(T)//2]),triFusionSansIfElse(T[len(T)//2:]))
    return T
```

```python
triFusionSansIfElse([3,1,4,1,5,9,2,6,5])
```

```python
import random
N=11111
triFusionSansIfElse([random.randint(0,N) for i in range(N)])
```

### Sans condition ?

Est-ce que l'on peut enlever les conditions de la boucle "A!=[] and B!=[]"

```python
def minMax(A,B):
    if A[0]<B[0]:
        return (A,B)
    return (B,A)
    
def fusionSansCondition(A,B):
    R=[]
    try:
        while True:
            (A,B)=minMax(A,B)
            R=R+[A[0]] 
            A=A[1:]
    except:
        return R+A+B
        

def triFusionSansCondition(T): 
    if len(T)>1:
        T=fusionSansCondition(triFusionSansCondition(T[0:len(T)//2]),triFusionSansCondition(T[len(T)//2:]))
    return T
```

### Sans try, ni while

On peut essayer de supprimer presque tout ...

```python
def minMax(A,B):
    if A[0]<B[0]:
        return (A,B)
    return (B,A)

def boucleInfini(Corps,Environnement):
    while True:
        Corps(Environnement)

def uneFusion(Env):
    (Env['A'],Env['B'])=minMax(Env['A'],Env['B'])
    Env['R']=Env['R']+[Env['A'][0]] 
    Env['A']=Env['A'][1:]
       
def fusionSansStructureControle(A,B):
    env={}
    env['R']=[]
    env['A']=A
    env['B']=B
    try:
        boucleInfini(uneFusion,env)
    except:
        return env['R']+env['A']+env['B']
        

def triFusionSansStructureControle(T): 
    if len(T)>1:
        T=fusionSansStructureControle(triFusionSansStructureControle(T[0:len(T)//2]),triFusionSansStructureControle(T[len(T)//2:]))
    return T
```

```python
def minMax(A,B):
    if A[0]<B[0]:
        return (A,B)
    return (B,A)

def boucleInfini(Corps,Environnement):
    while True:
        Corps(Environnement)

def executeSaufErreur(Corps,TraitantErreur,Environnement):
    try:
        Corps(Environnement)
    except:
        return TraitantErreur(Environnement)

def uneFusion(Env):
    (Env['A'],Env['B'])=minMax(Env['A'],Env['B'])
    Env['R']=Env['R']+[Env['A'][0]] 
    Env['A']=Env['A'][1:]  
    
def fusionFonctionnel(A,B):
    env={}
    env['R']=[]
    env['A']=A
    env['B']=B
    return executeSaufErreur(lambda E:boucleInfini(uneFusion,E), lambda E:E['R']+E['A']+E['B'],env)

def triFusionFonctionnel(T): 
    if len(T)>1:
        T=fusionFonctionnel(triFusionFonctionnel(T[0:len(T)//2]),triFusionFonctionnel(T[len(T)//2:]))
    return T
```

```python
triFusionFonctionnel([3,1,4,1,5,9,2,6,5])
```

```python
import random
N=11111
triFusionFonctionnel([random.randint(0,N) for i in range(N)])
```
