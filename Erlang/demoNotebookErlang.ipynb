## Partie 2 : pour 10 minutes de plus (*2)

*rappel : notebook = texte + code + exécution*

### Contexte éducatif

Hiérarchie de Bloom des objectifs éducationnels

* Niveau 1 : Connaissance (information)
* Niveau 2 : Compréhension (textuel)
* Niveau 3 : Application
* Niveau 4-A : Analyse (critique)
* Niveau 4-B : Evaluation
* Niveau 4-C : Synthèse (création)

Niveaux de compétences (E.N.)

* Niveau 1 : Connaitre
* Niveau 2 : Faire
* Niveau 3 : Enseigner

Compétences en programmation 

* Autour du code
  * lire un code
  * compléter un code
  * produire une code avec de l'aide
  * produire une code en suivant une méthode
  * modifier un code
  * produire une code sans aide
  * trouver une autre solution
  * corriger un code
* Autour de l'exécution
  * exécuter un code à la main
  * compiler/exécuter un code (par machine)
  * prévoir des jeux d'essais
  * comparer deux codes
  * choisir entre plusieurs solutions
  * évaluer un code
* Autour du texte
  * commenter un code
  * identifier et prouver des invariants
  * justifier des choix
  * expliquer un code
  * aider qlq'un à coder

Usages (avec exemples)

* coté enseignant
  * un cours (intéractif), ex. : [note](https://jupyterhub.u-ga.fr/user/bouhinde/notebooks/Github/notebooks/Python3/flottants.py.ipynb) sur les flottants en python
  * une démonstration, ex. [pense-bete](http://localhost:8888/notebooks/1_SyntaxeErlang.ipynb) "Syntaxe d'Erlang"
  * des élèments de correction d'examen, ex. [élèments de correction](https://swish.swi-prolog.org/p/ColoriageArcEnCielDeLaSuisse.swinb) pour colorier une carte en ProLog
  * une résolution, ex. [note de cours](http://localhost:8888/notebooks/Syracuse.ipynb) sur le calcul de la longueur maximale de suites de Syracuse en Erlang.
* coté étudiant
  * rendu de tp (compte-rendu) de ProLog, ex. :  [assez bien](https://swish.swi-prolog.org/p/CopyCREtudiant4.swinb), [assez bien](https://swish.swi-prolog.org/p/CopyCREtudiant3.swinb), [moyen](https://swish.swi-prolog.org/p/CopyCREtudiant2.swinb), [hors sujet](https://swish.swi-prolog.org/p/CopyCREtudiant.swinb)
  * travail dans un environnement/serveur contrôlé, ou travail collaboratif ou  (ex. : démo [jupyterhub](https://jupyterhub.u-ga.fr/) et [colaboratory](https://colab.research.google.com))
* exercices
  * exercices avec solution/élèments de correction (ex. : [démo](https://jupyterhub.u-ga.fr/user/bouhinde/notebooks/Python/Exemple%2C%20exercices.ipynb) avec Exemple exercice)
  * exercices avec validation (ex. : [démo](https://jupyterhub.u-ga.fr/user/bouhinde/notebooks/Python/Exemple%2C%20exercices.ipynb) avec Exemple exercice)
  
Démos
* des notebooks déjà rédigés
* des exemple d'exercice
* des exemples d'utilisation avec serveur/collaboratif : jupyterhub (uga : jupyterhub.u-ga.fr), colaboratory (google : colab.research.google.com)
